```python
import requests
import logging
import time
from datetime import datetime, timedelta
```

```python
# Get yesterday
now = datetime.now()
yesterday = now - timedelta(days=1)

# Set starting date as 1st Jan 2020
# starting_date = datetime(2020, 1, 1)

starting_date = datetime(2025, 11, 12)
```

```python
TIINGO_API_KEY = 'f63d6442dbb7d85699b6ed3863930f5b9b648c1f'
BASE_URL = "https://api.tiingo.com/tiingo/daily"
DELAY = 2


start_date = starting_date.date()
end_date = yesterday.date()
print(f"start_date: {start_date} | end_date: {end_date}")
```

```python
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

def fetch_price_data(tickers):
    data = []
    
    headers = {"Content-Type": "application/json"}
    for count, ticker in enumerate(tickers):
        try:
            url = f"{BASE_URL}/{ticker}/prices?token={TIINGO_API_KEY}&startDate={start_date}&endDate={end_date}"
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            price_data = response.json()
            if price_data:
                print(price_data)
                for daily_data in price_data:
                    data.append({
                        'date': daily_data.entry.get('date'),
                        'close': daily_data.entry.get('close'),
                        'high': daily_data.entry.get('high'),
                        'low': daily_data.entry.get('low'),
                        'open': daily_data.entry.get('open'),
                        'volume': daily_data.entry.get('volume'),
                        'adjClose': daily_data.entry.get('adjClose'),
                        'adjHigh': daily_data.entry.get('adjHigh'),
                        'adjLow': daily_data.entry.get('adjLow'),
                        'adjOpen': daily_data.entry.get('adjOpen'),
                        'adjVolume': daily_data.entry.get('adjVolume'),
                        'divCash': daily_data.entry.get('divCash')
                    })
                logger.info(f"{ticker}: price_data[0].get('close')")
            else:
                logger.warning(f"No data to fetch: {ticker}")
        except Exception as e:
            logger.warning(f"Error fetching {ticker}: {e}")
        time.sleep(DELAY)
    return data
```

```python
stocks = ["GOOG", "APPL"]
results = fetch_price_data(stocks)
for r in results:
    print("\n\nResults ____________\n\n")
    print(r)
```
