```python
def xy_split(data, time_step):
    df_x, df_y = [], []
    for i in range(len(data) - time_step - 1):
        x = data[i:(i + time_step)]
        df_x.append(x)
        df_y.append(data[i + time_step])
    return np.array(df_x), np.array(df_y)
```

```python
time_step = 100
X_train, y_train = xy_split(df_train, time_step)
X_test, y_test = xy_split(df_test, time_step)
```

- Test for Stationerity

```python
def test_stationarity(timeseries, col):

    rolmean = timeseries.rolling(7).mean()
    rolstd = timeseries.rolling(7).std()
    # Plot rolling statistics:
    plt.figure(figsize=[16,4])
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation of ' + col)
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft = adfuller(timeseries, autolag='AIC')
    # output for dft will give us without defining what the values are.
    # hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],
                       index=['Test Statistic',
                              'p-value',
                              'No. of lags used',
                              'Number of observations used'])

    for key, values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)

test_stationarity(df_nvidia_arima['Close'], 'Close')
test_stationarity(df_nvidia_arima['Open'], 'Open')
```

#### For Augmented Dickey Fuller (ADF) Test:

$H_0:$ Time series is non-stationary (i.e., variance is not constant over time)

$H_A:$ Time series is stationary

From the test above for closing price, p-value =0.977345, This is greater than .05, hence, we cannot reject the null hypothesis. We have to consider the time series to be non-stationary.

The test statistics is also above the critical values indicate  non-stationarity but with small degree.

```python
df_nvidia_arima['Close_diff'] = df_nvidia_arima['Close'] - df_nvidia_arima['Close'].shift(1)
```

```python
test_stationarity(df_nvidia_arima['Close_diff'].dropna(),'Close_diff')
```

**After applying first order diffrencing on close value the P-value is less than 0.05 so we can say this is statonary data**
