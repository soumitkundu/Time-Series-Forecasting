```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Download Hourly data for last 1 year
import yfinance as yf
nvidia = yf.Ticker("NVDA")
period = '2y'
df_nvidia_hourly = nvidia.history(period=period, interval='1h')
df_nvidia_hourly.to_csv('downloads/df_nvidia_stock_hourly_history.csv')
```

```python
df_nvidia_hourly = pd.read_csv('downloads/df_nvidia_stock_hourly_history.csv')
df_nvidia_hourly.head()
```

```python
df_nvidia_hourly_cleaned = df_nvidia_hourly[['Datetime', 'Open', 'High', 'Low', 'Close', 'Volume']].copy()
df_nvidia_hourly_cleaned.shape
```

```python
df_nvidia_hourly_cleaned.dtypes
```

```python
df_nvidia_hourly_cleaned['Datetime'] = pd.to_datetime(df_nvidia_hourly_cleaned['Datetime'], utc=True)
df_nvidia_hourly_cleaned['Datetime'] = df_nvidia_hourly_cleaned['Datetime'].dt.tz_localize(None)
# df_nvidia_hourly_cleaned['Datetime'] = df_nvidia_hourly_cleaned['Datetime'].dt.tz_convert('Asia/Kolkata')
df_nvidia_hourly_cleaned.dtypes
```

```python
df_nvidia_hourly_cleaned.head()
```

```python
df_nvidia_hourly_cleaned['Date'] = df_nvidia_hourly_cleaned['Datetime'].dt.date
df_nvidia_hourly_cleaned['Time'] = df_nvidia_hourly_cleaned['Datetime'].dt.time
df_nvidia_hourly_cleaned['Day'] = df_nvidia_hourly_cleaned['Datetime'].dt.day
df_nvidia_hourly_cleaned['Day of Week'] = df_nvidia_hourly_cleaned['Datetime'].dt.dayofweek
df_nvidia_hourly_cleaned['Month'] = df_nvidia_hourly_cleaned['Datetime'].dt.month
df_nvidia_hourly_cleaned['Year'] = df_nvidia_hourly_cleaned['Datetime'].dt.year

df_nvidia_hourly_cleaned.head()
```

```python
df_nvidia_hourly_cleaned.to_csv('downloads/nvidia_stock_hourly_data.csv')
```
